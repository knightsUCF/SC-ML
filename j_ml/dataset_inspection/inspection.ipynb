```python
import os
import re
```

```python
if not os.path.exists(new_dir):
    os.mkdir(new_dir)
```

```python
 # processed_dge folder contains all new defined datasets
 # get all shapes from processed_dge files
 shape_list = []
 for root, dirs, files in os.walk("/home/jay/Documents/projects/todo_files/ml_final_pro/datasets/processed_dge", topdown=False):
     for name in files:
        file_dir = os.path.join(root, name)
        tmp_df = pd.read_csv(file_dir)
        tmp_groups = tmp_df.groupby('target_id')
        cluster_id_list = []
        number_of_obs_list = []
        for group in tmp_groups:
            cluster_id_list.append(group[0])
            number_of_obs_list.append(len(group[1]))
        plt.bar(cluster_id_list, number_of_obs_list)
        plt.xlabel('cluster id')
        plt.ylabel('number of observations')
        plt.title(name)
        new_dir = '../../documents/images'
        new_dir = os.path.join(new_dir, name)
        new_dir = re.sub('.csv','.jpg', new_dir)
        plt.savefig(new_dir)
        plt.show()
```
