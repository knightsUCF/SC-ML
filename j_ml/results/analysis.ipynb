```python
def read_accuracy(in_file_dir):
    results_list = []
    for root, dirs, files in os.walk(in_file_dir, topdown=False):
        for name in files:
            file_dir = os.path.join(root, name)
            results_list.append(file_dir)

    results_list = sorted(results_list)
    list_accuracy = []
    for file_dir in results_list:
        with open(file_dir, 'r') as f:
            real_infor = []
            lines = f.readlines()
            for line in lines:
                if line != "\n":
                    real_infor.append(line.strip())
            list_accuracy.append(round(float(real_infor[-1])*100,2))
    return list_accuracy
```
